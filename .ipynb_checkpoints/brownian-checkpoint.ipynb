```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

A **brownian motion process** (also known as a Wiener process) is a stochastic process characterized by the property

$$ B(t+\Delta t)-B(t)\sim N(0,\sqrt{\Delta t})=\sqrt{\Delta t}N(0,1) $$

and that $B$ is continuous and $B(0)=0$.

```python
def simulate_brownian_process(num_sims, N, T):
    deltaT = T / N
    t = np.linspace(0, T, N+1)
    X = np.c_[np.zeros((num_sims, 1)), np.random.randn(num_sims, N)]
    return t, np.cumsum(np.sqrt(deltaT) * X, axis=1)
```

```python
def plot_bm(num_sims, N):
    T = 2
    t, B = simulate_brownian_process(num_sims, N, T)
    fig = plt.figure(figsize=(7,7))
    ax = fig.add_subplot(111)    
    ax.plot(t, B.T)
    ax.plot(t, 2 * np.sqrt(t), "black")
    ax.set_xlabel('t',fontsize=20)
    ax.set_ylabel('W(t)',fontsize=20)
    plt.tight_layout()
plot_bm(1000, 10)
```

Arithmetic brownian motion is given by the stochastic differential equation (SDE)

$$ dX_t = \mu dt + \sigma dB_t $$

```python
def simulate_arithmetic_brownian_motion(num_sims,N,t0,B0,T,mu,sigma):
    deltaT = T / N
    t = np.linspace(t0, t0+T, N+1)
    X = np.random.randn(num_sims, N)
    X = np.c_[np.ones((num_sims, 1)) * B0, mu * deltaT + sigma * np.sqrt(deltaT) * X]
    X = np.cumsum(X, axis=1)
    return t, X
```

```python
def plot_arith_bm(num_sims,N,t0,B0,T,mu,sigma):
    t, X = simulate_arithmetic_brownian_motion(num_sims,N,t0,B0,T,mu,sigma)
    fig = plt.figure(figsize=(7,7))
    ax = fig.add_subplot(111)    
    ax.plot(t, X.T)
    ax.set_xlabel('t',fontsize=20)
    ax.set_ylabel('W(t)',fontsize=20)
    plt.tight_layout()
```

```python
num_sims = 100
N = 500
t0 = 0
T = 2
B0 = 0
mu = 1
sigma = .5

plot_arith_bm(num_sims, N, t0, B0, T, mu, sigma)
```
