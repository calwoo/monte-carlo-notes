```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Now we'll use euler's method for ODEs with stochastic differential equations (SDEs). The idea here is that we will instead use euler's method on simulated trajectories, which we will integrate over.

Let us look at the standard SDE with a Wiener process giving our stochasticity:
$$ df_t = a(t,f_t)dt + b(t, f_t)dB_t $$
Recall here that $B_{t+T}-B_t \sim \sqrt{T}N(0,1)$.

```python
def stochastic_euler_integration(t0,T,f0,a,b,N,num_sims):
    """
    Stochastic euler method for numerical integration of an SDE.
    INPUT
        t0,T = as in deterministic version
        f0 = f(t0)
        a, b = the functions in the SDE above
        N = number of time steps
        num_sims = number of simulated paths
    OUTPUT
        t = times over which the paths occur
        f = values of the trajectory f starting at f0 at t0
    """
    deltaT = T / N
    t = np.linspace(t0, t0+T, N+1)
    # initialize f
    f = np.zeros((num_sims, N+1))
    # generate stochasticity
    stoch = np.random.randn(num_sims, N) * np.sqrt(deltaT)
    f[:,0] = f0
    
    for n in range(N):
        f[:,n+1] = f[:,n] + a(t[n], f[:,n]) * deltaT + b(t[n], f[:,n]) * stoch[:,n]
    return t, f
```

As an example, lets try to simulate the solutions to the geometric Brownian motion
$$ dS_t = \mu S_t dt + \sigma S_t dB_t $$
We note that the fact that this SDE can be solved analytically gives us a nice comparison. In particular, the paths should match those simulated from the solution
$$ S_t = S_0\exp\left(\left(\mu-\frac{1}{2}\sigma^2\right)t + \sigma B_t\right) $$

```python
def gbm_sde(mu, sigma):
    def a(t,st):
        return mu * st
    def b(t,st):
        return sigma * st
    return a, b
```

```python
def gbm():
    S0    = 100   # initial asset price
    t0    = 0.0   # initial date
    mu    = 0.3   # expected return
    sigma = .1    # volatility
    T     = 10.   # length of time interval in years
    N     = 20    # number of integration steps
    M     = 1000  # number of trajectories
    
    a, b = gbm_sde(mu, sigma)
    t, f = stochastic_euler_integration(t0,T,S0,a,b,N,M)
    
    fig = plt.figure(figsize=(7,5))
    plt.plot(t, f.T)
    
gbm()
```

```python
def gbm_exact():
    S0    = 100   # initial asset price
    t0    = 0.0   # initial date
    mu    = 0.3   # expected return
    sigma = .1    # volatility
    T     = 10.   # length of time interval in years
    N     = 20    # number of integration steps
    M     = 1000  # number of trajectories
    
    deltaT = T / N
    t = np.linspace(t0,t0+T,N+1)
    S = np.zeros((M,N+1))
    S[:,0] = S0
    stoch = np.random.randn(M,N)
    for n in range(N):
        S[:,n+1] = S[:,n] * np.exp((mu-0.5*sigma**2)*deltaT + sigma*np.sqrt(deltaT)*stoch[:,n])
        
    fig = plt.figure(figsize=(7,5))
    plt.plot(t, S.T)
    
gbm_exact()
```

So it seems that the stochastic Euler method is giving the correct solutions.
